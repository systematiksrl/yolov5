### **Appunti**
`--weights` e' il modello gia' allenato da caricare.
E' possibile allenare un modello piu' volte cambiandogli i dati od aggiungendoglieli.  

`--data` e' il file path dove e' presente il file dataset.yaml con cui e' possibile configurare dove prendere i dati per l'allenamento del modello,
quante e quali classi ci sono presenti nel dataset fornito.
